```python
#Removing Outliers Using  IQR

columns = ['Weight','Length1','Length2','Length3','Height','Width']

for column in columns:
    iqr = np.percentile(fis[column],75) -  np.percentile(fis[column],25)
    upper_limit =  np.percentile(fis[column],75) + (1.5 * (iqr))
    lower_limit =  np.percentile(fis[column],25) - (1.5 * (iqr))
    print(iqr, upper_limit, lower_limit)
    fis.drop(fis[fis[column]>upper_limit].index,inplace = True)
    fis.drop(fis[fis[column]<lower_limit].index,inplace = True)
    
    
```

# PREDICTIVE ANALYSIS

```python
#Splitting the data
from sklearn.model_selection import train_test_split
features = [col for col in fish_final.columns if col != 'Weight']

X_train, X_test, y_train, y_test = train_test_split(fish_final[features], fish_final['Weight'], test_size=.2, random_state=22)
X_train.shape, X_test.shape
```

```python
fish_final.columns
```

```python
features
```

```python
X_train
```

```python
#Normalising the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
linearregression = LinearRegression()                                    
linearregression.fit(X_train, y_train) 
```

```python
linearregression.coef_
```

```python
linearregression.intercept_
```

```python
y_predict = linearregression.predict(X_test)
```

```python
y_predict
```

```python
pred = linearregression.predict(X_test)
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

# Mean Absolute Error
print("\nMean Absolute Error:", mean_absolute_error(y_test, pred))
# RMSE
print("\nRMSE:", mean_squared_error(y_test, pred)**0.5)
# R2 Squared:
print("\nR2 Squared - Score:", r2_score(y_test, pred))
# Training Score
print("\nTraining Score:", linearregression.score(X_train, y_train))
# Testing score
print("\nTesting score:",linearregression.score(X_test, y_test))
```
